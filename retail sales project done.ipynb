```python
import pandas as pd
df = pd.read_csv('retail sales dataset.csv')
```

```python
print(df.isnull().sum())
```

```python
print(df.info())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
print(df.describe())
```

```python
print(df['Product Category'].value_counts())
```

```python
print(df.groupby('Gender')['Total Amount'].mean())
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
monthly_sales = df.groupby(['Year', 'Month'])['Total Amount'].sum()
print(monthly_sales)
```

```python
import matplotlib.pyplot as plt
monthly_sales.plot(kind='line')
plt.title('Monthly Sales Trends')
plt.ylabel('Total Sales')
plt.show()
```

```python
top_products = df['Product Category'].value_counts().head(5)
print(top_products)
```

```python
df['Age'].hist(bins=20)
plt.title('Customer Age Distribution')
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title('Customer Gender Distribution')
plt.show()
```

```python
df.groupby('Product Category')['Total Amount'].sum().sort_values().plot(kind='barh')
plt.title('Total Sales by Product Category')
plt.show()
```

```python
df.groupby(pd.cut(df['Age'], bins=5))['Total Amount'].mean().plot(kind='bar')
plt.title('Average Purchase by Age Group')
plt.show()
```
